## Analog to Digital
Everything **real** is analog. All of our technology is digital (discrete times and discrete numbers). So, all analog signals, or electromagnetic waves, must be converted to digital information to be worked with on computers.

```python
bits = 2
levels = np.linspace(-1, 1, 2**bits)
levels
```

```python
bits = 12
levels = 2**bits
intervals = levels - 1
intrvl = 3.3 / intervals
ADCindex = 2733
ADCinput = intrvl * ADCindex
ADCinput
```

```python
bits = 14
levels = 2**bits
intervals = levels - 1
intrvl = 5 / intervals
voltage = 3.3
ADCout = voltage / intrvl
ADCout
#BUT!!! IT SHOULD BE AN INTEGER
```
