We had multiple failed write attempts for adding the Raspberry Pi OS to the board. We even switched SD cards, but only after trying Balena Etcher and the Windows diskpart utility (which actually still gave an error, but having seen the SD card listed in the file explorer we moved on), did the OS successfully write onto the card. After that, it was simply a matter of connecting the Raspberry Pi to power, then to the router, and then signing into it with SSH. The test used to ensure remote access was working was an execution of the "echo" command.
