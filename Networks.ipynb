## Introduction to Networks
First public demonstration of the internet was in 1972, 54 years ago (as of 2026).

There used to be a literal link between the two callers, an occupied circuit. Circuits had to be switched by a human operator. This system was inefficient, and led to the invention of packet switching.

Packet switching allows for the original message to be divided into packets. Each packet could take a different path. A single connection could fail, but there are multiple redundant paths for packets to still arrive at their destination.

Local area networks are networks contained with in a small area (home, office, restaurant). The LAN is protected by the firewall. Typical devices are computers and printers, routers and switches.

The number system used is hexadecimal, or a shorthand version of binary. It is base 16, from 0 to 9 and A to F. 0 (HexDec) is 0000 (binary) and A is 1010.

Converting binary to decimal is straightforward. Each spot, from right to left, represents 2 to the power of the index. So 8 bits would represent 128, 64, 32, 16, 8, 4, 2, and 1. The binary number "10011011" can be converted to decimal adding the values from where a 1 can be found. So "10011011" in decimal is "128+16+8+2+1" or 155. Converting a number from decimal to binary is relatively simple. Take the starting number, like 225, and divide it by 2. If there is a remainder, add a 1 to the binary version and move one space left. Take the number that fit (so 112) and divide that by 2 as well. There is no remainder, so add a 0 and move another space left.
*There is a simpler way, though.* Take the number, 225, and take the largest exponent that fits into 225, 128, and subtract. The 128 space becomes a 1, then take the difference, 97, and see if the next space to the right, 64, fits. It does, so subtract and add a 1 to the 64 space. If any space's number doesn't fit, add a 0 and skip. [bla](https://www.youtube.com)
